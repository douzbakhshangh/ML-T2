```python
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import learning_curve
import matplotlib.pyplot as plt
import pandas as pd
```

```python
MODEL = LinearRegression(normalize=True)
x = np.random.rand(200, 1) * 20
y=x**4+x**3+x**2+x
y = y + np.random.normal(scale=3, size=y.shape)
x = x.reshape(-1, 1)
y = y.reshape(-1, 1)
```

```python
split = int(x.shape[0] * 0.8)
x_train = x[:split]
y_train = y[:split]
x_test = x[split:]
y_test = y[split:]
```

```python
MODEL.fit(x_train, y_train)
x_line = np.arange(-2.5, 22.5, 0.1).reshape(-1, 1)
y_line = MODEL.predict(x_line)
plt.plot(x_train, y_train, 'b.')
plt.plot(x_test, y_test, 'c.')
plt.plot(x_line, y_line, 'm--',label='Part1\ndegree1')
plt.xlabel('x-axis',fontname='Comic Sans MS', fontsize=17,color='aqua')
plt.ylabel('y-axis',fontname='Comic Sans MS', fontsize=17,color='aqua')
plt.title('over/under fitting',fontname='Comic Sans MS', fontsize=20,color='navy')
plt.grid()
plt.legend(loc=2, borderpad=1.5,facecolor='lightyellow',edgecolor='aqua')
plt.savefig('exercise2.png', dpi=300)
```

```python
MSE_train = []
MSE_test = []
h_train = MODEL.predict(x_train)
mse_train = np.mean((y_train - h_train) ** 2) / 2
h_test = MODEL.predict(x_test)
mse_test = np.mean((y_test - h_test) ** 2) / 2
MSE_train.append(mse_train)
MSE_test.append(mse_test)
print('MSE_train=',MSE_train)
print('mean mse_train=',np.mean(MSE_train))
print('-----------------------------------------------------------')
print('MSE_test=',MSE_test)
print('mean mse_test=',np.mean(MSE_test))
```
